```python
corpus='''
The cat sat lazily on the windowsill, watching the world pass by, while a gentle breeze swept through the park, carrying the scent of blooming flowers.
She carefully packed her suitcase, excited for the adventure that awaited, as the coffee shop down the street bustled with people, yet felt cozy and calm.
He looked up at the sky, marveling at the vibrant colors of the setting sun, while the train whizzed by in the distance, leaving a cloud of dust in its wake.
An old book sat forgotten on the shelf, its pages yellowed with age, as the sound of laughter echoed through the halls where children played outside. 
The mountains stood tall and majestic, their peaks touching the clouds, as she took a deep breath before stepping onto the stage, her heart racing with anticipation.
'''
```

```python
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
documents=sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
from nltk.tokenize import word_tokenize
```

```python
words=word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
for word in words:
    print(word+'--------->'+stemmer.stem(word))
```

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
reg_stemmer.stem('eating')

```

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+'--------->'+snowballstemmer.stem(word))
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
lematizer=WordNetLemmatizer()
nltk.download('wordnet')
nltk.download('punkt')
```

```python
for word in words:
    print(word+'--------->'+lematizer.lemmatize(word))
```
